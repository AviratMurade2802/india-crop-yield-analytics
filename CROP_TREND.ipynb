# Crop Production, Area and Yield Trends (2006–2011)

This notebook analyzes **production, area and yield trends** for major crops over 2006–07 to 2010–11 using `datafile (2).csv`.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (9, 4)
```

```python
df2 = pd.read_csv("datafile (2).csv")
df2.head()
```

```python
df2.info()
```

```python
df2.columns
```

```python
# strip spaces and make lowercase with underscores
df2.columns = [c.strip().lower().replace(" ", "_") for c in df2.columns]
df2.head()
```

```python
# strip spaces and make lowercase with underscores
df2.columns = [c.strip().lower().replace(" ", "_") for c in df2.columns]
df2.head()
```

```python
# Identify production/area/yield columns by prefix
prod_cols = [c for c in df2.columns if c.startswith("production_")]
area_cols = [c for c in df2.columns if c.startswith("area_")]
yield_cols = [c for c in df2.columns if c.startswith("yield_")]

prod_cols, area_cols, yield_cols
```

```python
# Melt each metric to long and then merge
prod_long = df2.melt(
    id_vars=["crop"],
    value_vars=prod_cols,
    var_name="metric_year",
    value_name="production"
)

area_long = df2.melt(
    id_vars=["crop"],
    value_vars=area_cols,
    var_name="metric_year",
    value_name="area"
)

yield_long = df2.melt(
    id_vars=["crop"],
    value_vars=yield_cols,
    var_name="metric_year",
    value_name="yield_val"
)

# Ensure same order
prod_long = prod_long.sort_values(["crop", "metric_year"]).reset_index(drop=True)
area_long = area_long.sort_values(["crop", "metric_year"]).reset_index(drop=True)
yield_long = yield_long.sort_values(["crop", "metric_year"]).reset_index(drop=True)

# Combine into one dataframe
ts = prod_long.copy()
ts["area"] = area_long["area"]
ts["yield_val"] = yield_long["yield_val"]

ts.head()
```

```python
# metric_year like 'production_2006-07' -> get the year part
ts["year"] = ts["metric_year"].str.extract(r"(\d{4}-\d{2})")
ts = ts.drop(columns=["metric_year"])

ts.head()
```

```python
ts["year"].value_counts().sort_index()
```

```python
ts["crop"].value_counts().head()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=ts, x="year", y="yield_val", hue="crop", marker="o")
plt.title("Yield trends by crop (2006–07 to 2010–11)")
plt.xticks(rotation=45)
plt.ylabel("Yield (index or units as per source)")
plt.tight_layout()
plt.show()
```

```python
selected_crops = ["Rice", "Wheat", "Maize", "Total Foodgrains"]  # adjust based on df
mask = ts["crop"].isin(selected_crops)

plt.figure(figsize=(10, 5))
sns.lineplot(data=ts[mask], x="year", y="production", hue="crop", marker="o")
plt.title("Production trends for selected crops")
plt.xticks(rotation=45)
plt.ylabel("Production (units as per source)")
plt.tight_layout()
plt.show()
```

```python
summary = ts.groupby("crop")["yield_val"].agg(["mean", "min", "max"]).reset_index()
summary.sort_values("mean", ascending=False).head(10)
```

```python
# pivot to crop x year for yield
pivot_yield = ts.pivot_table(
    index="crop", columns="year", values="yield_val"
)

pivot_yield.head()
```

```python
# compute growth from first to last available year
years_sorted = sorted(pivot_yield.columns)
first_year, last_year = years_sorted[0], years_sorted[-1]
pivot_yield["growth"] = pivot_yield[last_year] - pivot_yield[first_year]

pivot_yield.sort_values("growth", ascending=False).head(10)
```
