# Crop Varieties, Seasons and Recommended Zones

This notebook explores crop **varieties**, their **season/duration**, and **recommended zones** using `datafile (3).csv'


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (9, 4)

df3 = pd.read_csv("datafile (3).csv")
df3.head()
```

```python
df3.columns = ["crop", "variety", "season_duration", "recommended_zone", "extra"]

df3["crop"] = df3["crop"].astype(str).str.strip()
df3["variety"] = df3["variety"].astype(str).str.strip()
df3["season_duration"] = df3["season_duration"].astype(str).str.strip()
df3["recommended_zone"] = df3["recommended_zone"].astype(str).str.strip()

df3 = df3.drop(columns=["extra"])
df3.head()
```

```python
df3.info()
```

```python
df3["crop"].value_counts()
```

```python
varieties_per_crop = df3.groupby("crop")["variety"].nunique().reset_index(name="n_varieties")
varieties_per_crop.sort_values("n_varieties", ascending=False)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=varieties_per_crop.sort_values("n_varieties", ascending=False),
            x="crop", y="n_varieties")
plt.title("Number of listed varieties per crop")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
df3["season_duration"].value_counts().head(20)
```

```python
import re
import numpy as np

def parse_duration(text):
    text = text.strip()
    if not text or text.upper() == "NA":
        return np.nan
    # if contains digits
    nums = re.findall(r"\d+", text)
    if nums:
        nums = list(map(int, nums))
        return sum(nums) / len(nums)
    # for words like Medium, Mid-early we keep NaN or map manually
    return np.nan

df3["duration_days"] = df3["season_duration"].apply(parse_duration)
df3[["season_duration", "duration_days"]].head(10)
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df3.dropna(subset=["duration_days"]),
            x="crop", y="duration_days")
plt.xticks(rotation=45, ha="right")
plt.title("Variety duration (days) by crop")
plt.tight_layout()
plt.show()
```

```python
df3["recommended_zone"].head(10)
```

```python
# count how often each state name appears (quick heuristic)
states = ["Andhra Pradesh", "Tamil Nadu", "Karnataka", "Gujarat", "Maharashtra",
          "Uttar Pradesh", "Bihar", "Rajasthan", "Madhya Pradesh", "West Bengal",
          "Orissa", "Odisha", "Punjab", "Haryana", "Assam", "Jharkhand",
          "Chhattisgarh", "Uttarakhand", "Jammu", "Kashmir", "Manipur", "Meghalaya"]

for s in states:
    df3[s] = df3["recommended_zone"].str.contains(s, case=False, na=False).astype(int)

zone_counts = df3[states].sum().sort_values(ascending=False)
zone_counts.head(10)
```

```python
plt.figure(figsize=(8, 4))
zone_counts.head(10).plot(kind="bar")
plt.title("Top states in recommended zones (by count of varieties)")
plt.ylabel("Number of varieties recommended")
plt.tight_layout()
plt.show()
```
